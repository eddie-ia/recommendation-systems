```python
file_paths = [
    'data/data_canal_digital.csv',
    'data/data_consumos_tarjetas.csv',
    'data/data_productos.csv',
    'data/data_ubicacion_geografica.csv'
    ]

def load_dataframes(file_paths):
    dfs = []
    for path in file_paths:
        df = pd.read_csv(path,sep=',')
        dfs.append(df)
    return dfs
```

```python
dfs = load_dataframes(file_paths)
```

```python
dfs[1].head()
```

```python
import matplotlib.pyplot as plt
```

```python
for idx, df in enumerate(dfs):
    df_name = file_paths[idx].split('/')[-1].replace('.csv', '')

    # ---- Parte 1: Gráfica general por DataFrame ----
    unique_counts = df.nunique()
    null_counts = df.isnull().sum()

    fig, ax = plt.subplots(figsize=(10, 5))
    ax.bar(unique_counts.index, unique_counts.values, label='Valores únicos', alpha=0.7)
    ax.bar(null_counts.index, null_counts.values, label='Valores nulos', alpha=0.7)

    ax.set_title(f"Resumen columnas: {df_name}")
    ax.set_ylabel("Cantidad")
    ax.set_xticks(range(len(df.columns)))
    ax.set_xticklabels(df.columns, rotation=45, ha='right')
    ax.legend()
    plt.tight_layout()
    plt.show()

    # ---- Parte 2: Gráfica + Detalle por columna ----
    for col in df.columns:
        value_counts = df[col].value_counts(dropna=False).head(10)

        # Imprimir detalle
        print(f"\n📌 Columna: {col} — {df_name}")
        for val, count in value_counts.items():
            print(f"{val} → {count} ocurrencias")

        # Generar gráfico de barras para los top 10
        fig, ax = plt.subplots(figsize=(8, 4))
        value_counts.plot(kind='bar', ax=ax)
        ax.set_title(f"{df_name} - {col} (top 10)")
        ax.set_ylabel("Frecuencia")
        ax.set_xlabel("Valor")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
```
